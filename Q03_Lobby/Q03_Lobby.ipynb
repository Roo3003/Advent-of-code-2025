# Part 1

```python
total_joltage = 0

for bank in banks:

    # Get the tens digit
    # - - - - - - - - - - - - - -

    # Look for left-most index of largest observed digit
    seen_digits = {}

    for i,digit in enumerate(bank):
        if digit not in seen_digits:
            seen_digits[digit] = i
    
    max_obs_digit = max(seen_digits)

    # Need to handle differently if the largest digit is on the very end
    if seen_digits[max_obs_digit] == len(bank)-1:
        new_dict = seen_digits.copy()
        seen_digits.pop(max_obs_digit)
        max_obs_digit = max(seen_digits)

    
    # Get the units digit
    # - - - - - - - - - - - - -
    
    max_digit_idx = seen_digits[max_obs_digit]
    remaining_bank = bank[max_digit_idx+1:]

    max_second_digit = 0
    for digit in remaining_bank:
        max_second_digit = max(max_second_digit,int(digit))
    max_second_digit = str(max_second_digit)


    total_joltage += int(max_obs_digit+max_second_digit)    


total_joltage
```

# Part 2

```python
# Generalise to X batteries being turned on for a bank of N batteries
 
# process needs to be...

# let A := X

# 1) Cut off the final N-(A-1) batteries
# 2) Find the left-most index of the largest digit and turn it on
# 3) Disregard everything including & to the left of that digit for determining the remaining digits
# 4) Set A := A - 1
# 5) Repeat 1-4 until A = 0 at step 3

# e.g. X=4, N=10, bank=4723472309 

# cut off last 3

# 4723472

# leftmost largest digit is the 7 at index 2
# disregard everything to the left

# cut off last 2

# x7|234723

# leftmost largest digit is 7 at index 6
# disregard everything to the left

# cut off last 1

# x7xxx7|230

# leftmost largest digit is 3 at index 8
# disregard everything to the left

# cut off last 0

# x7xxx7x3|09

# leftmost largest digit is 9 at index 10
# disregard everything to the left

# x7xxx7x3x9

# => joltage = 7739
```

```python
total_joltage = 0

batteries_to_turn_on = 12


for bank in banks:
    #print('\n')
    joltage = ''

    left_boundary = 0

    for num_batteries_to_cut_off in range(batteries_to_turn_on,0,-1):
        if num_batteries_to_cut_off == 1:
            cut_off_bank = bank[left_boundary:]
        else:
            cut_off_bank = bank[left_boundary:-num_batteries_to_cut_off+1]
        #print(cut_off_bank, num_batteries_to_cut_off)

        # Get leftmost largest digit
        seen_digits = {}

        for i,digit in enumerate(cut_off_bank):
            if digit not in seen_digits:
                seen_digits[digit] = i+left_boundary+1
        
        largest_digit = max(seen_digits)
        left_boundary = seen_digits[largest_digit]

        joltage += largest_digit
        #print(joltage)
    
    total_joltage += int(joltage)

total_joltage
```
