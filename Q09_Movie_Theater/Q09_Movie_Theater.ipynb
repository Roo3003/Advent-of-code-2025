```python
puzzle = open('puzzle.txt', 'r').read()
example = open('example.txt', 'r').read()
```

```python
input = puzzle

tiles = input.split('\n')
```

# Part 1

```python
max_area = 0

def get_area(tile1,tile2):
    x_len = abs(tile1[0]-tile2[0])+1
    y_len = abs(tile1[1]-tile2[1])+1
    return x_len*y_len


for tile_id_1 in range(len(tiles)):
    for tile_id_2 in range(tile_id_1):
        tile1 = ((int(tiles[tile_id_1].split(',')[0])), (int(tiles[tile_id_1].split(',')[1])))
        tile2 = ((int(tiles[tile_id_2].split(',')[0])), (int(tiles[tile_id_2].split(',')[1])))

        this_area = get_area(tile1,tile2)
        max_area = max(max_area,this_area)

max_area
```

# Part 2

```python
import itertools

vertices = [tuple(map(int, l.split(','))) for l in input.split('\n')]
wrapped_vertices = vertices+[vertices[0]]
vertices

max_area = 0

# Iterate through every possible rectangle
for rectangle in itertools.combinations(vertices,2):
    # Unpack rectangle
    (x1,y1),(x2,y2) = rectangle[0],rectangle[1]
    
    # Re-define the rectangle so we always use the bottom left and top right corners
    lowest_x, largest_x = min(x1,x2), max(x1,x2)
    lowest_y, largest_y = min(y1,y2), max(y1,y2)

    #print(f'Rectangle {(lowest_x,lowest_y)},{(largest_x,largest_y)}')

    # Check whether rectangle is valid or not

    # Iterate through all edges of the perimeter
    for edge_start,edge_end in itertools.pairwise(wrapped_vertices):

        edge_start_x, edge_end_x = edge_start[0], edge_end[0]
        edge_start_y, edge_end_y = edge_start[1], edge_end[1]

        # Does the edge's maximum x index steer clear of our rectangle?
        max_x_clear = max(edge_start_x,edge_end_x) <= lowest_x

        # Does the edge's minimum x index steer clear of our rectangle?
        min_x_clear = min(edge_start_x,edge_end_x) >= largest_x

        # Does the edge's maximum y index steer clear of our rectangle?
        max_y_clear = max(edge_start_y,edge_end_y) <= lowest_y

        # Does the edge's minimum y index steer clear of our rectangle?
        min_y_clear = min(edge_start_y,edge_end_y) >= largest_y

        # Is the rectangle invalid?
        # If all conditions are False (i.e. there is no way to get around the line) then rectangle is invalid
        if not(max_x_clear or max_y_clear or min_x_clear or min_y_clear):
            break
    # If we didn't trigger the break condition (i.e. the rectangle is valid), update the area
    else:
        max_area = max(max_area, get_area(rectangle[0],rectangle[1]))
        #print(f'New max area = {max_area}')

max_area
```
