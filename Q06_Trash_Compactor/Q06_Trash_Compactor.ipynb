```python
puzzle = open('puzzle.txt', 'r').read()
example = open('example.txt', 'r').read()
```

```python
input = puzzle

# Parse input

worksheet = input
rows = worksheet.split('\n')
symbols = [x for x in rows[-1].split(' ') if x != '']
nums = []
for row in rows:
    nums.append([int(x) for x in row.split(' ') if x not in ['','*','+']])
nums = nums[:-1]
```

# Part 1

```python
total = 0

for problem,sign in enumerate(symbols):
    match sign:
        case '*':
            count = 1
            for num_row in nums:
                count *= num_row[problem]
        
        case '+':
            count = 0
            for num_row in nums:
                count += num_row[problem]
    
    total += count

total
```

# Part 2

```python
import numpy as np

# Parse input

# We can repeat the part 1 method if we can transpose the input set of numbers
# So, we put the string characters into an np matrix, transpose it, then concat the new strings back together
# Then feed those results back into our solution for part 1


worksheet = input
rows = worksheet.split('\n')
symbols = [x for x in rows[-1].split(' ') if x != '']
str_nums = rows[:-1]

# Initialise matrix
str_nums_mat = np.empty(shape=(len(str_nums),len(str_nums[0])), dtype='object')

for row,num_string in enumerate(str_nums):
    for col,s in enumerate(num_string):
        str_nums_mat[row][col] = s

str_nums_mat = str_nums_mat.T
# Concat strings back together
nums = []

concat_str = ''
for i,row in enumerate(str_nums_mat):
    if (not all([x == ' ' for x in row])) and i != (len(str_nums_mat)-1):
        concat_str += ''.join(row)
    elif i == (len(str_nums_mat)-1):
        concat_str += ''.join(row)
        nums.append(concat_str)
        concat_str = ''
    else:
        nums.append(concat_str)
        concat_str = ''
    concat_str += ' '

int_nums = []
for row in nums:
    int_nums.append([int(x) for x in row.split(' ') if x != ''])
```

```python
total = 0

for problem,sign in enumerate(symbols):
    match sign:
        case '*':
            count = 1
            for num in int_nums[problem]:
                count *= num
        case '+':
            count = 0
            for num in int_nums[problem]:
                count += num
    
    total += count

total
```
