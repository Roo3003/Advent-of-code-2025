# Part 1

```python
fewer_than_four_count = 0

rows, cols = len(diagram), len(diagram[0])

def check_spot(spot):
    paper_count = 0

    #example_str = '\n'
    for adj_row in [-1,0,1]:
        for adj_col in [-1,0,1]:

            row_to_check = int(spot.real + adj_row)
            col_to_check = int(spot.imag + adj_col)

            # Skip self & out of bounds
            if (adj_col==0 and adj_row==0) or (row_to_check < 0) or (col_to_check < 0) or (row_to_check > rows-1) or (col_to_check > cols-1):
                #example_str += 'X'
                continue
            
            #print(row_to_check,col_to_check)
            adj_check = diagram[row_to_check][col_to_check]
            if adj_check == '@':
                paper_count += 1

            #example_str += adj_check
        #example_str += '\n'
    
    #print(example_str)
    
    return paper_count

for row in range(rows):
    for col in range(cols):
        if check_spot(row + col*1j) < 4 and diagram[row][col] == '@':
            #print(f'row {row} col {col} Above has < 4 paper!')
            fewer_than_four_count += 1

fewer_than_four_count
```

# Part 2

```python
removed_count = 0

rows, cols = len(diagram), len(diagram[0])

def check_spot(spot,diagram):
    paper_count = 0

    #example_str = '\n'
    for adj_row in [-1,0,1]:
        for adj_col in [-1,0,1]:

            row_to_check = int(spot.real + adj_row)
            col_to_check = int(spot.imag + adj_col)

            # Skip self & out of bounds
            if (adj_col==0 and adj_row==0) or (row_to_check < 0) or (col_to_check < 0) or (row_to_check > rows-1) or (col_to_check > cols-1):
                #example_str += 'X'
                continue
            
            #print(row_to_check,col_to_check)
            adj_check = diagram[row_to_check][col_to_check]
            if adj_check == '@':
                paper_count += 1

            #example_str += adj_check
        #example_str += '\n'
    
    #print(example_str)
    
    return paper_count

i = 0

cur_diagram = diagram

while(True):
    i += 1

    #print(cur_diagram)

    removed_this_time = 0

    next_diagram = []
    for row in range(rows):
        new_row = ''
        for col in range(cols):
            if check_spot(row + col*1j,cur_diagram) < 4 and cur_diagram[row][col] == '@':
                #print(f'row {row} col {col} Above has < 4 paper!')
                new_row += '.'
                removed_this_time += 1
            else:
                new_row += cur_diagram[row][col]
        next_diagram.append(new_row)

    removed_count += removed_this_time

    cur_diagram = next_diagram
    if i > 100 or removed_this_time == 0:
        break

removed_count
```
