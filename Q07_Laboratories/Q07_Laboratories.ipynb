# Part 1

```python
# Slightly scuffed; thought the original prompt was how many unique beams there were coming out of splitters, not just the number of splitters.
# This works but could probably be improved by just doing DFS while keeping a record of unique splitter locations

starting_row, starting_col = 0, manifold[0].find('S')
num_rows, num_cols = len(manifold), len(manifold[0])

beam_locs = set([(starting_row,starting_col)])
beams_to_eval = set([(starting_row,starting_col)])

seen_beams = []

splitter_locs = set()

while(beams_to_eval):

    for b in beams_to_eval:
        seen_beams.append(b)

    next_beam_set = set()

    for beam in beams_to_eval:
        beam_row, beam_col = beam[0], beam[1]
        for ahead_row in range(beam_row, num_rows):
            if manifold[ahead_row][beam_col] == '^':
                splitter_locs.add((ahead_row,beam_col))
                if (ahead_row, beam_col-1) not in beam_locs:
                    next_beam_set.add((ahead_row, beam_col-1))
                    beam_locs.add((ahead_row, beam_col-1))

                if (ahead_row, beam_col+1) not in beam_locs:
                    next_beam_set.add((ahead_row, beam_col+1))
                    beam_locs.add((ahead_row, beam_col+1))
                break

            else:
                beam_locs.add((ahead_row,beam_col))
    
    beams_to_eval = next_beam_set

print(len(splitter_locs))
```

# Part 2

```python
# DFS with memorisation

def quantum_beams(row,col,mem):
    '''
    Given a starting beam position (row,col), returns the number
    of different timelines the beam ends up on (i.e. how many paths it can take)
    '''

    for ahead_row in range(row,num_rows):
        count = 0
        if manifold[ahead_row][col] == '^':
            if mem[(ahead_row,col)] != -1:
                #print(f'Seen splitter {(ahead_row,col)} before: {mem[(ahead_row,col)]} sub-problems')
                return mem[(ahead_row,col)]
            
            count_left = quantum_beams(ahead_row,col-1,mem)
            count_right = quantum_beams(ahead_row,col+1,mem)
            count = count_left + count_right
            #print(f'Exhausted splitter {(ahead_row,col)}: {count} sub-problems ({count_left} left, {count_right} right)')
            mem[(ahead_row,col)] = count
            break
        elif ahead_row == num_rows-1:
            return count + 1

    return count

def init_quantum_beams():
    splitter_map = {}

    for row in range(len(manifold)):
        for col in range(len(manifold[0])):
            if manifold[row][col] == '^':
                splitter_map[(row,col)] = -1
    
    return splitter_map

quantum_beams(starting_row,starting_col,init_quantum_beams())
```
