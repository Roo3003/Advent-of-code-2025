```python
puzzle = open('puzzle.txt', 'r').read()
example = open('example.txt', 'r').read()
```

```python
input = puzzle
input = input.split('\n')
#input
```

```python
rotations = input

num = 50
zero_count = 0

for rot in rotations:
    match rot[0]:
        case 'L': num = (num - int(rot[1:])) % 100
        case 'R': num = (num + int(rot[1:])) % 100
    if num == 0:
        zero_count += 1

print(zero_count)
```

```python
def get_clicks(num, rot, dir):

    click_count = 0
    movement_remaining = rot

    match dir:
        case 'L':
            buffer = num
        case 'R':
            buffer = 100-num
    
    while(movement_remaining > 0):
        if movement_remaining >= 100:
            click_count += 1
            movement_remaining -= 100
        elif movement_remaining >= buffer and num != 0:
            click_count += 1
            return click_count
        else:
            return click_count

get_clicks(25, 125, 'L')
```
