# Part 1

```python
def invalid_check(num):
    n = len(str(num))

    if n % 2 != 0:
        return 0
    
    first_half = str(num)[:int(n/2)]
    second_half = str(num)[int(n/2):]

    if first_half == second_half:
        return int(num)
    else:
        return 0
```

```python
input = puzzle

invalid_count = 0

for num_range in input.split(','):
    lower, upper = num_range.split('-')
    
    for i in range(int(lower),int(upper)+1):
        invalid_count += invalid_check(i)

invalid_count
```

```python
def invalid_check(num):

    n = len(str(num))

    # get divisors of n

    divisors_of_n = []

    for i in range(int(n//2)):
        if n % (i+1) == 0:
            divisors_of_n.append(i+1)
    
    for seq_len in divisors_of_n:
        substrings = [str(num)[i:i+seq_len] for i in range(0,n,seq_len)]
        
        if substrings.count(substrings[0]) == len(substrings):
            return num
        
    return 0


invalid_check(111)
```
