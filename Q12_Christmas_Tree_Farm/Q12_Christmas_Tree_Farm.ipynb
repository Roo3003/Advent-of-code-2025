```python
puzzle = open('puzzle.txt', 'r').read()
example = open('example.txt', 'r').read()
```

```python
input = puzzle

present_info = input.split('\n\n')
```

# Part 1

```python
# A general solution seems intractable to find. We instead look at the input and examine the area available
# vs the area that the presents will take to fill

shapes, regions = present_info[:-1], present_info[-1].split('\n')

shape_map = {}

shape_space = {}

for shape in shapes:
    shape_idx = int(shape[0])
    s = [[s] for s in shapes[shape_idx][3:].split('\n')]

    hash_count = 0
    for row in s:
        for rowstr in row:
            for char in rowstr:
                if char == '#':
                    hash_count += 1

    shape_map[shape_idx] = s
    shape_space[shape_idx] = hash_count

presents_can_likely_fit = 0

for region in regions:
    area_desc, presents_required = region.split(':')

    # Extract area
    length,width = map(int,area_desc.split('x'))
    area = length*width

    # Extract presents required
    present_area_required = 0
    for present_idx,present_num in enumerate(presents_required.strip().split(' ')):
        present_num = int(present_num)

        present_area_required += present_num*shape_space[present_idx]

    
    if present_area_required < area:
        #print(f'{area-present_area_required} free area')

        # After some expermentation, all the cases which have potentially sufficient area have at least
        # 300 free space leftover if you could optimally pack the presents with no 'holes'

        # This is a large margin so we simply return the number of cases where there is sufficient area

        if area-present_area_required < 300:
            print('Close case')
        else:

            presents_can_likely_fit += 1


presents_can_likely_fit
```
